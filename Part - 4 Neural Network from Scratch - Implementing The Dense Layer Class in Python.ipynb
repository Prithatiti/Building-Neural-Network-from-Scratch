### Dense Layer Class

Note: *numpy.random.randn* Return a sample (or samples) from the “standard normal” distribution.

```python
class Layer_Dense:
    # Layer initialization
    def __init__(self, n_inputs, n_neurons):
        # Initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    # Forward pass
    def forward(self, batch_inputs):
        # Calculate the output values from inputs, weights and biases
        self.output = np.dot(batch_inputs, self.weights) + self.biases

# Create Dataset
X, Y = spiral_data(samples = 100, classes = 3)

# Create Dense Layer with 2 input features and 3 output values
dense_1 = Layer_Dense(2, 3)

# Perform a forward pass of our training data through this layer
dense_1.forward(X)

# Let's see the output of the first few samples
print(dense_1.output[: 5])
```
